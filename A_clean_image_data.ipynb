```python
import os
from PIL import Image
import shutil

data_path = "sports_data"
clean_path = "clean_sports_data"
os.makedirs(clean_path, exist_ok=True)
image_size = (224, 224)

for sport in os.listdir(data_path):
    sport_folder = os.path.join(data_path, sport)
    if os.path.isdir(sport_folder):
      new_sport_folder = os.path.join(clean_path, sport)
      os.makedirs(new_sport_folder, exist_ok=True)

      for img_file in os.listdir(sport_folder):
        img_path = os.path.join(sport_folder, img_file)
        try:
          with Image.open(img_path) as img:
            img = img.convert("RGB")
            img = img.resize(image_size)
            img.save(os.path.join(new_sport_folder, img_file))
        except Exception as e:
          print(f"Skipped: {img_file}, Reason: {e}")
```
